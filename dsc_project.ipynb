```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"/content/student_ml_dataset.csv")
```

```python
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()

# Example of a count plot for a categorical variable
sns.countplot(df['Hours_Studied'])
plt.ylabel('Count')
plt.title('Hours Studied')
plt.show()
```

```python
plt.scatter(df['Age'], df['Physics_Marks'])
plt.xlabel('Age')
plt.ylabel('Physics Marks')
plt.title('Age vs. Physics Score')
plt.show()

plt.scatter(df['Age'], df['Math_Marks'])
plt.xlabel('Age')
plt.ylabel('Maths Marks')
plt.title('Age vs. Maths Score')
plt.show()

plt.scatter(df['Age'], df['Chemistry_Marks'])
plt.xlabel('Age')
plt.ylabel('Chemistry Marks')
plt.title('Age vs. Chemistry Score')
plt.show()

# Example of a correlation matrix to explore correlations between numerical variables
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Hours_Studied', 'IQ']])
plt.title('Pair Plot of Age, IQ, and Study Hours')
plt.show()
```
